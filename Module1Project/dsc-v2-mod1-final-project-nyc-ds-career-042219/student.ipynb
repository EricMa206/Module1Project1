* **view** - Has been viewed
* **condition** - How good the condition is ( Overall )

* **zipcode** - zip
* **lat** - Latitude coordinate
* **long** - Longitude coordinate
* **sqft_living15** - The square footage of interior housing living space for the nearest 15 neighbors
* **sqft_lot15** - The square footage of the land lots of the nearest 15 neighbors

# Data Cleaning

```python
# waterfront and yr_renovated are missing too many values. Removed data index.
kc_df.drop(['id','waterfront', 'yr_renovated',], axis=1, inplace=True)
```

```python
kc_df.isna().sum()
```

```python
# view is missing some rows but few enough that I can just drop those rows
kc_df.dropna(axis=0, inplace=True)
```

```python
# There are '?' in the sqft_basment data. I'll keep that in mind.
print(kc_df.loc[kc_df['sqft_basement']=='?', 'sqft_basement'].count())
```

```python
kc_df['zipcode'].unique().size
```

```python
kc_df['zipcode'].hist();
```

```python
# Split the zipcodes into bins where close by zipcodes are generally grouped together. 
bins = [98000, 98020, 98040, 98060, 98100, 98120, 98160, 98200]
bins_zip = pd.cut(kc_df['zipcode'], bins)
zip_dummies = pd.get_dummies(bins_zip, prefix="zip", drop_first=True)
zip_dummies.head()
```

```python
# Create zipcode bin dummy variables.
kc_df = pd.concat([kc_df, zip_dummies], axis=1)
kc_df.drop(['zipcode'], axis=1, inplace = True)
kc_df
```

### The Kitchen Sink model

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

outcome = 'price'
pred_sum = "+".join(kc_pred.columns)
formula = outcome + "~" + pred_sum
```

---Not a great result. Its high considering I took the kitchen sink approach. Lets go the opposite direction. I know the 'sqft_living' is correlated with price.

### The staircase model

```python
outcome = 'price'
formula = outcome + "~" + 'sqft_living'
model = ols(formula= formula, data=kc_df).fit()
model.summary()
```

--- Add 'bedrooms' and more layers.

```python
outcome = 'price'
formula = 'price ~ sqft_living+sqft_lot+bedrooms+yr_built+condition'
model = ols(formula= formula, data=kc_df).fit()
model.summary()
```
